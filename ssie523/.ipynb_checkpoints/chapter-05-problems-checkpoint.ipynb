```python
#Exercise 5.4
def model(xinit, yinit, itr):
    
    xresults = [xinit]
    yresults = [yinit]
    
    for each in range(0, itr):
        lastx = xresults[-1]
        lasty = yresults[-1]

        nextx = lastx + 0.1*(lastx - lastx*lasty)
        nexty = lasty + 0.1*(lasty - lastx*lasty)
        
        xresults.append(nextx)
        yresults.append(nexty)

    results = {}
    
    results['yresults'] = yresults
    results['xresults'] = xresults
    
    return results

def plot(lower_lim, upper_lim, step, model, itr, prob):
    
    param_space = np.arange(lower_lim, upper_lim, step)
    param_list = []
    for each_a in param_space:
        for each_b in param_space:
            param_list.append({'param_a': each_a, 'param_b':each_b})

    plt.figure(figsize=(15, 5))     
    for i in range(len(param_list)):
        run1 = model(xinit=float(param_list[i]['param_a']), 
                     yinit=float(param_list[i]['param_b']), 
                     itr=itr)

        plt.plot(run1['xresults'], run1['yresults'])
        plt.xlabel('X-Values')
        plt.ylabel('Y-Values')
    plt.title(prob)
    plt.axis([0, upper_lim, 0, upper_lim])
    plt.show()
    
plot(lower_lim=0.025, upper_lim=5.0, step=0.5, model=model, itr=30, prob='Problem 5.4')
```

```python
#Exercise 5.8

g = nx.DiGraph()

for x in range(6):
    for y in range(6):
        g.add_edge((x, y), ((x*y) % 6, x))
a
ccs = [each for each in nx.connected_components(g.to_undirected())]
n = len(ccs)
w = np.ceil(sqrt(n))
h = np.ceil(n / w)

for e in range(n):
    plt.subplot(h, w, e+1)
    nx.draw(nx.subgraph(g, ccs[e]), with_labels=True)
plt.show()
```

```python
#Exercise 5.12
A = np.array([[1, -1, 0], 
              [-1, -3, 1], 
              [0, 1, 1]])

eigen_stuffs = np.linalg.eig(A)
eigen_stuffs
```

```python
print('Dominant Eigenvalue: {:.5f}'.format(eigen_stuffs[0][np.argmax([abs(each) for each in eigen_stuffs[0]])]))
```

```python
print('Dominant Eigenvector: ')
print(eigen_stuffs[1][:, np.argmax([abs(each) for each in eigen_stuffs[0]])].reshape(3, 1))
```

```python
np.linalg.matrix_power(A, 10).dot(np.array([1, 0.1, 0.4])).reshape(3, 1)
```

```python
#Exercise  5.16
x, y = symbols('x y')

F1 = x + 2*x*(1-x) - x*y - x
F2 = y + 2*y*(1-y) - x*y - y

#Step 1: Compute eq. points.
solve([F1, F2], [x, y])
```

```python
#Step 2: Compute Jacobian matrix.
J = Matrix([[F1.diff(x), F1.diff(y)], 
            [F2.diff(x), F2.diff(y)]])
J
```

```python
# Step 3: Sub eq. point into Jacobian matrix.
J_0 = J.subs([(x, 2/3), (y, 2/3)])
J_0
```

```python
# Step 4: Determine eigenvalues of J_0
J_0.eigenvals()
```

```python
# Step 4: Determine eigenvalues of J_O
J_0.eigenvects()
```

```python
#Exercise 5.17

F1 = x*y - x
F2 = y*(x-1) - y

#Step 1: Compute eq. points.
solve([F1, F2], [x, y])
```

```python
#Step 2: Compute Jacobian matrix.
J = Matrix([[F1.diff(x), F1.diff(y)], 
            [F2.diff(x), F2.diff(y)]])
J
```

```python
# Step 3: Evalulate Jacobian Matrix around eq. point 1.
J_first = J.subs([(x, 0), (y, 0)])
J_first
```

```python
# Step 3a: Compute Eiganvalues for Jacobian matrix with eq. point 1. 
J_first.eigenvects()
```

```python
# Step 3b: Evalulate Jacobian Matrix around eq. point 2.
J_second = J.subs([(x, 2), (y, 1)])
J_second
```

```python
# Step 3c: Compute Eiganvalues for Jacobian matrix with eq. point 2. 
J_second.eigenvects()
```

#### Discussion RE: Exercise 5.17.

For the first eq. point, $(x=0, y=0)$ the dominant eigenvalue was $\lambda = -2$. Since $|\lambda| > 1$, the model is unstable around the point $(x=0, y=0)$. For the second eq. point, $(x=2, y=1)$ the dominant eigenvalue was $\lambda = \sqrt{2}$. Since $|\lambda| > 1$, the model is unstable around the point $(x=2, y=1)$.

```python
# Modeling Exercise from Lectures

def knowledge(sinit, ainit, xinit, itr, r_a, r_s, n):
    xresults = [xinit]
    aresults = [ainit]
    sresults = [sinit]
    
    results = {}
    
    for each in range(0, itr):
        x_last = xresults[-1]
        s_last = sresults[-1]
        a_last = aresults[-1]
        
        x_new = x_last + a_last * (1/n)*(1 - s_last)
        a_new = a_last - x_last * a_last + r_a * s_last
        s_new = s_last + r_s * s_last*(1 - s_last)
        
        xresults.append(x_new)
        sresults.append(s_new)
        aresults.append(a_new)
    
    results['x'] = xresults
    results['s'] = sresults
    results['a'] = aresults
    
    [x_new for each in range(0, itr)]
    [s_new for each in range(0, itr)]
    [a_new for each in range(0, itr)]

    return results
```

```python
run = knowledge(sinit=0.01, ainit=1, xinit=0.05, itr=30, r_a=0.5, r_s=0.35, n=30)
plt.plot(run['x'], label='Knowledge')
plt.plot(run['s'], label='Stress')
plt.plot(run['a'], label='Attention')
plt.legend()
plt.show()
```

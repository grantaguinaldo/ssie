```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p_param = 0.25
Pvalues, Nvalues = np.meshgrid(np.arange(0, 100, 0.1), 
                               np.arange(0, 100, 0.1))

Pdot = (p_param-Pvalues)*Nvalues*Pvalues
Ndot = -Nvalues*Pvalues

plt.streamplot(Pvalues, Nvalues, Pdot, Ndot)
plt.ylabel('Popularity', fontsize=14)
plt.xlabel('Novelty', fontsize=14)
plt.show()
```

```python
from sympy import *
init_printing()
```

```python
P, I, Pc, a, b, c = symbols('P I P_c, a b c')
Pdot = a * (Pc - P) * I * P - c
Idot = -b * I * P

#Find Equlibirum Points
solve([Pdot, Idot], (P, I))

#(P, 0) should not be a zero point.
#(0, I) people don't know about your song.
# Not valid
```

```python
P, I, Pc, a, b, c = symbols('P I P_c, a b c')
Pdot = a * (Pc - P) * I * P - c*P
Idot = -b * I * P

#Find Equlibirum Points
solve([Pdot, Idot], (P, I))

#(0, 0) should not be a zero point.
#(0, I) people don't know about your song.
```

```python
solve(Pdot, (P, I))
#No change in Popularity, P (Point in Phase Space where Pdot=0)
```

```python
solve(Idot, (P, I))
#No change in Interest, I (Point in Phase Space where Idot=0)
```

```python
P, I = np.meshgrid(np.arange(0, 2, 0.1), 
                               np.arange(0, 2, 0.1))

a, b, Pc = 1, 1, 1

Pdot = a * (Pc - P) * I * P - c*P
Idot = -b * I * P


plt.streamplot(P, I, Pdot, Idot)
plt.ylabel('Popularity', fontsize=14)
plt.xlabel('Novelty', fontsize=14)
plt.show()
```

```python
x_init = 0.05
a_init = 0.075
s_init = 0.01
num_classes = 30
x = [x_init]
a = [a_init]
s = [s_init]

for each in range(0, num_classes):
    
    nx = x[-1] + a[-1]*(1/num_classes) * (1 - s[-1])
    na = a[-1] - x[-1] * a[-1] + 0.1 * s[-1]
    #ns = s[-1] + 0.1 * s[-1]
    
    x.append(nx)
    a.append(na)
    #s.append(ns)
    
plt.plot(x, label='Knowledge')
plt.plot(a, label='Attention')
#plt.plot(s, label='Stress')
plt.legend()
plt.show()
```

```python
def attention(x_init, a_init, s_init, 
              a_rate, s_rate, num_classes, itr):
    
    results_x = [x_init]
    results_a = [a_init]
    results_s = [s_init]
    
    for each in range(0, itr):
        last_x = results_x[-1]
        last_a = results_a[-1]
        last_s = results_s[-1]
    
        nx = last_x + last_s * (1/num_classes) * (1-last_s)
        na = last_a - last_x * last_a + a_rate * last_s 
        ns = last_s + s_rate * last_s * (1-last_s)
        
        results_x.append(nx)
        results_a.append(na)
        results_s.append(ns)
        
    results = {}
    
    results['results_x'] = results_x
    results['results_a'] = results_a
    results['results_s'] = results_s
    
    return results   
```

```python
run = attention(x_init=0.05, s_init=0.05, a_init=1, a_rate=0.1, s_rate=0.3, num_classes=30, itr=30)
plt.figure(figsize=(15, 5))
plt.plot(run['results_x'], label = 'Knowledge')
plt.plot(run['results_a'], label='Attention')
plt.plot(run['results_s'], label='Stress')
plt.xlim(0, 30)
plt.ylim(0, 1)
plt.xlabel('Time Steps', fontsize=14)
plt.ylabel('Signal (arb. units)', fontsize=14)
plt.legend()
plt.show()
```

```python
def model(x_init, a_init, class_num, alpha, itr):
    results_x = [x_init]
    results_a = [a_init]
    
    for each in range(0, itr):
        #Include Noise in the amount that someone pays attention.
        n_new = results_x[-1] + results_a[-1] * (1/class_num)
        
        # Attention does not depend on Knowledge.
        #a_new = alpha * results_a[-1]
        a_new = results_a[-1] - results_x[-1] * results_a[-1]
        
        results_x.append(n_new)
        results_a.append(a_new)
        
    results = {}
    results['results_x'] = results_x
    results['results_a'] = results_a
    
    return results

run = model(x_init=0.05, a_init=1, alpha = 0.95, class_num=30, itr=30)
plt.plot(run['results_x'], label='Knowledge')
plt.plot(run['results_a'], label='Attention')
plt.legend()
plt.show()
```

```python
def model(x_init, a_init, s_init, class_num, alpha, itr):
    results_x = [x_init]
    results_a = [a_init]
    results_s = [s_init]
    
    for each in range(0, itr):
        n_new = results_x[-1] + results_a[-1] * (1/class_num) * (1-results_s[-1])
        a_new = results_a[-1] - results_x[-1] * results_a[-1] + 0.24*results_s[-1]
        s_new = results_s[-1] + 0.1 * results_s[-1] 
        
        results_x.append(n_new)
        results_a.append(a_new)
        results_s.append(s_new)
        
    results = {}
    results['results_x'] = results_x
    results['results_a'] = results_a
    results['results_s'] = results_s
    
    return results

run = model(x_init=0.05, a_init=1, s_init=0.1, alpha = 0.9, class_num=30, itr=30)
plt.plot(run['results_x'], label='Knowledge')
plt.plot(run['results_a'], label='Attention')
plt.plot(run['results_s'], label='Stress')
plt.legend()
plt.show()
```
